```python
# Convert the date column into a pandas datetime object
df["Date"] = pd.to_datetime(df["Date"])
```

```python
# Extract the Year, Month, Quarter
df["year"] = df["Date"].dt.year
df["month"] = df["Date"].dt.month
df["month_name"] = df["Date"].dt.month_name()
df["quarter"] = df["Date"].dt.quarter

df.head(2)
```

```python
#Group Customer age
def age_group(x):
  if x <= 25:
    return "<25 Young Adult"
  elif x <= 40:
    return "25-40 Adult"
  elif x <= 50:
    return "41-50 Old Adult"
  else:
    return ">=51 Elders"

# Appy function to the data
df["age_group"] = df["Customer_Age"].apply(age_group)
df.head(1)
```

```python
#cost, Revenue and pofit calculation and adding them as new columns
df["cost"] = df["Quantity"]*df["Unit_Cost"]
df["revenue"] = df["Quantity"]*df["Unit_Price"]
df["profit"] = df["revenue"] - df["cost"]

df.head(2)
```

```python
# Grouping revenue into Profit/loss
def PandL(x):
  if x > 0:
    return "Profit"
  elif x < 0:
    return "Loss"
  else:
    None

df["Profit_label"] = df["profit"].apply(PandL)
df.head(2)
```

# Numerical Visualization - Quality, Cost, Revenue and Profit - Subplot

## Bivarant Analysis - Involves the relationship between two variables

```python
#create the rolls and columns
#Catergorical Variables
fig,axs = plt.subplots(nrows=2, ncols=3, figsize=(15,10))

customer_profit = df.groupby("Customer")["profit"].sum().reset_index()
sns.barplot(x="Customer", data=customer_profit, y="profit", ax=axs[0,0])
axs[0,0].set_title("Total profit by customer type")

Sales_Person_profit = df.groupby("Sales Person")["profit"].sum().reset_index()
top_5 = df["Sales Person"].value_counts().head(5)
sns.barplot(x="Sales Person", data=Sales_Person_profit, y="profit",  ax=axs[0,1], order=top_5.index)
axs[0,1].set_title("Total profit by Sales Person")

Product_Category_profit = df.groupby("Product_Category")["profit"].sum().reset_index()
sns.barplot(x="Product_Category", data=Product_Category_profit, y="profit", ax=axs[0,2])
axs[0,2].set_title("Total profit by Product Category")

Payment_Option_profit = df.groupby("Payment Option")["profit"].sum().reset_index()
sns.barplot(x="Payment Option", data=Payment_Option_profit, y="profit", ax=axs[1,0])
axs[1,0].set_title("Total profit by Payment Option")

Sub_Category_profit = df.groupby("Sub_Category")["profit"].sum().reset_index()
top_3 = df["Sub_Category"].value_counts().head(3)
sns.barplot(x="Sub_Category", data=Sub_Category_profit, y="profit", ax=axs[1,1],  order=top_3.index)
axs[1,1].set_title("Total profit by Sub_Category")

customer_Age_profit = df.groupby("Customer_Age")["profit"].sum().reset_index()
top_5 = df["Customer_Age"].value_counts().head(5)
sns.barplot(x="Customer_Age", data=customer_Age_profit, y="profit", ax=axs[1,2], order=top_5.index)
axs[1,2].set_title("Total profit by customer Age")

```
